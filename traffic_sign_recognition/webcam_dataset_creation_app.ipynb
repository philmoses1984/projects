```python
#Projekt_11_09_24
import cv2
import os
from PIL import Image

# Function to capture images from webc    am and save them as PNG
def capture_images(save_folder):
    # Check if save folder exists, if not create it
    if not os.path.exists(save_folder):
        os.makedirs(save_folder)

    # Open the webcam
    cap = cv2.VideoCapture(0)

    # Set the image size for capturing (optional: OpenCV will capture larger and we'll resize it)
    img_width, img_height = 224, 224
    img_counter = 0  # To count and name the captured images

    print("Press 'Space' to capture image, 'Esc' to exit.")

    while True:
        # Capture frame-by-frame
        ret, frame = cap.read()

        # Display the frame in a window
        cv2.imshow("Press 'Space' to capture, 'Esc' to exit.", frame)

        # Wait for user input
        key = cv2.waitKey(1) & 0xFF

        if key == 27:  # ESC key to exit
            print("Exiting...")
            break
        elif key == 32:  # SPACE key to capture image
            img_name = f"image_{img_counter}.png"  # Save as PNG
            img_path = os.path.join(save_folder, img_name)
            
            # Convert the captured image (from OpenCV) to a PIL image
            pil_image = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))

            # Resize the image to 224x224 using LANCZOS resampling
            pil_image = pil_image.resize((img_width, img_height), Image.Resampling.LANCZOS)

            # Save the image as PNG
            pil_image.save(img_path, format='PNG')
            print(f"Image saved: {img_path}")
            img_counter += 1

    # Release the webca m and close all OpenCV windows
    cap.release()
    cv2.destroyAllWindows()

# Main function
if __name__ == "__main__":  
    # Define the folder where images will be saved
    save_folder = r"C:\Users\phili\Documents\Python\Fotos_für_Projekt_11_09_2024\Buch"  # Replace this with your desired folder path
    capture_images(save_folder)
```
